Title: Predict the percentage of a student based on the no. of study hours.

Question: What will be predicted score if a student studies for 9.25 hrs/day?

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Importing Dataset
df = pd.read_csv("http://bit.ly/w-data")
```

```python
df.head(10)
```

```python
df.describe()
```

```python
# Dropping rows having null values
df.dropna(axis=0, how="any", inplace = True)
```

```python
# Visualising Data
df.plot(x="Hours", y="Scores", style='o')
plt.title('Hours study and percentage scored relation graph')
plt.xlabel('Hours studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
# Divide the data into attributes (inputs) and "labels" (outputs)
x = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

```python
# Split data into training and test data
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=0)
```

```python
# Train the algorithm
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Testing Algorithm
y_predict = regressor.predict(x_test)

# Comparing actual values with the predicted values
df2 = pd.DataFrame({"Predicted Values": y_predict, "Actual Values": y_test, "Difference": y_predict-y_test})
print(df2)
```

```python
# Eva;uate the model
from sklearn import metrics
print("Mean Absolute Error:", metrics.mean_absolute_error(y_test, y_predict))
```

```python
# Now Predict required data
z = float(input())
regressor.predict([[z]])
```
