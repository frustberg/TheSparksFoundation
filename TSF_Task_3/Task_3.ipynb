# Handling Outliners

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[-((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

# Correlation

```python
df.corr()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),cmap="YlGnBu", annot=True)
```

# Visualization

```python
df.City.value_counts().nlargest(20).plot(kind='bar', figsize=(10,5))
plt.rc_context({'axes.edgecolor':'#581845', 'xtick.color':'black', 'ytick.color':'black', 'figure.facecolor':'#f6eee2'})
plt.title("Total Business in the cities")
plt.ylabel('Business')
plt.xlabel('Cities')
```

```python
df.State.value_counts().nlargest(20).plot(kind='bar', figsize=(10,5))
plt.title("Total Business in the States")
plt.ylabel('Business')
plt.xlabel('Cities')
```

```python
sns.jointplot(x='Profit', y='Sales', data = df)
plt.show()
```

```python
# To modify the length of the graph
plt.title("Quantity vs Profit", fontweight = 'bold', fontsize = 20)
plt.xlabel("", fontsize = 15, fontweight = 'bold')
plt.ylabel("", fontsize = 15, fontweight = 'bold')

sns.barplot(x = 'Quantity', y = 'Profit', data = df)
plt.show()
```

```python
sns.catplot(x='Category', y='Quantity', hue = "Sub-Category", kind='box', height = 5, aspect = 2.5, data = df)
plt.title("Category vs Quantity", fontweight = 'bold', fontsize = 20)
plt.xlabel("Category", fontsize = 15, fontweight = 'bold')
plt.ylabel("Quantity", fontsize = 15, fontweight = 'bold')
plt.show()
```

```python
plt.title("Discount vs Profit", fontweight = 'bold', fontsize = 20)
plt.xlabel("", fontsize = 15, fontweight = 'bold')
plt.ylabel("", fontsize = 15, fontweight = 'bold')

sns.barplot(x = 'Discount', y = 'Profit', data = df[df.Profit > 0])
plt.show()
```

A discount of 0.1 to 0.4 units is ideal to gain a profit

```python
plt.title("Discount vs Sales", fontweight = 'bold', fontsize = 20)
plt.xlabel("", fontsize = 15, fontweight = 'bold')
plt.ylabel("", fontsize = 15, fontweight = 'bold')

sns.barplot(x = 'Discount', y = 'Sales', data = df[df.Profit > 0])
plt.show()
```

A discount of 0.3 units results in a highest amount of sales.

```python
sns.catplot(x='Category', y='Quantity', kind='boxen', data = df)
```

Office supplies gains the highest amount of Products

```python
plt.title("Category vs Sales", fontweight = 'bold', fontsize = 20)
plt.xlabel("", fontsize = 15, fontweight = 'bold')
plt.ylabel("", fontsize = 15, fontweight = 'bold')

sns.barplot(x = 'Category', y = 'Sales', data = df)
plt.show()
```

Furniture products have the minimum amount if sales whereas Office Suppliers have the least.

```python
sns.catplot(x='Sales', y='Category', hue = "Sub-Category", kind = "bar", height=6 ,aspect=1.5, data = df)
plt.grid()
plt.show()
```

```python
plt.title("Category vs Profit", fontweight = 'bold', fontsize = 20)
plt.xlabel("", fontsize = 15, fontweight = 'bold')
plt.ylabel("", fontsize = 15, fontweight = 'bold')

sns.barplot(x = 'Profit', y = 'Category', data = df)
plt.show()
```

Technological products gain the highest profit among all.

```python
sns.violinplot(x='Region', y='Profit', data = df)
```
